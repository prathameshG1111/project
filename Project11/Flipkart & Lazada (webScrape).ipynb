```python
#---------Flipkart product data-----------------

import pandas as pd
import numpy as np
import time
import requests
import selenium
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from bs4 import BeautifulSoup


def getUrl(key):
    key = key.replace(' ','+')
    temp = 'https://www.flipkart.com/search?q={}'
    url = temp.format(key)
    return url

def get_product(card):
    # Fetching product_name
    try:
        product_name = card.find('a','IRpwTa').text
    except:
        product_name = 'Not Mentioned'
        
    # Fetching product_price
    try:
        product_price = card.find('div','_30jeq3').text
        product_price = float(''.join(filter(str.isdigit, product_price)))
    except:
        product_price = 'Not Mentioned'
        
    # Fetching product_desc
    try:
        product_desc = card.find('div','_3Ay6Sb').text
    except:
        product_desc = 'Not Mentioned'

    # Fetching product_img_url
    try:
        product_img_url = card.find('img','_2r_T1I')['src']
    except:
        product_img_url = 'Not Mentioned'
        
    # Fetching product_buy_link
    try:
        product_buy_link = card.find('a','_2UzuFa')
        product_buy_link = 'https://www.flipkart.com'+ product_buy_link['href'] 
    except:
        product_buy_link = 'Not Mentioned'
    
    product_compete_info = (product_name, product_price, product_desc, product_img_url, product_buy_link)
    return product_compete_info

def clean_price(price):
    # Check if the price is 'Not Mentioned'
    if isinstance(price, str) and price == 'Not Mentioned':
        return np.nan  # Return NaN for missing prices
    else:
        try:
            return float(price.replace('â‚¹', '').replace(',', ''))
        except AttributeError:
            return price  

def main(product):
    all_products = []
    url = getUrl(product)
    driver = webdriver.Firefox()
    driver.get(url)
    driver.maximize_window()
    res = requests.get(url)
    soup = BeautifulSoup(res.content, 'html.parser')
    all_divs = soup.find_all('div', '_1xHGtK _373qXS')
    for i in all_divs:
        temp = get_product(i)
        all_products.append(temp)
    col = ['product_name', 'product_price', 'product_desc', 'product_img_url', 'product_buy_link']
    flipcart_product_data = pd.DataFrame(all_products, columns=col)
    
    # Clean and convert product_price column to numeric for sorting
    flipcart_product_data['product_price'] = flipcart_product_data['product_price'].apply(clean_price)
    
    # Sort DataFrame by product_price in descending order
    flipcart_product_data = flipcart_product_data.sort_values(by='product_price', ascending=False)
    flipcart_product_data.to_csv(r'C:\\iGurus\\FlipKart (web Scrap)\\flipcart_product_data(jeans).csv', encoding='utf-8', index=False)
```

```python
#--------------- Lazada Product data------------------

import pandas as pd
import numpy as np
import time
import requests
from selenium import webdriver
from bs4 import BeautifulSoup

def getUrl(key):
    key = key.replace(' ','+')
    temp = 'https://www.lazada.com.my/tag/{}'
    url = temp.format(key)
    return url

def get_products(card):
    # Fetching product_name
    try:
        product_name = card.find('div', 'RfADt').text
    except:
        product_name = 'Not Mentioned'

    # Fetching product_price
    try:
        product_price = card.find('span', 'ooOxS').text
        product_price = clean_price(product_price)
    except:
        product_price = 'Not Mentioned'

    # Fetching product_desc
    try:
        product_desc = card.find('span', 'IcOsH').text
    except:
        product_desc = 'Not Mentioned'

    # Fetching product_img_url
    try:
        product_img_url = card.find('div', 'picture-wrapper').find('img')['src']
    except:
        product_img_url = 'Not Mentioned'

    # Fetching product_buy_link
    try:
        product_buy_link = card.find('div', 'RfADt').find('a')['href']
    except:
        product_buy_link = 'Not Mentioned'

    all_product_info = (product_name, product_price, product_desc, product_img_url, product_buy_link)
    return all_product_info

def clean_price(price):
    # Check if the price is 'Not Mentioned'
    if isinstance(price, str) and price == 'Not Mentioned':
        return np.nan  # Return NaN for missing prices
    else:
        # Remove 'RM' and any non-numeric characters, then convert to float
        price = price.replace('RM', '').replace(',', '').strip()
        return float(price)

def main(product):
    all_products = []
    url = getUrl(product)
    driver = webdriver.Firefox()
    driver.get(url)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    cards = soup.find_all('div', 'Bm3ON')
    for i in cards:
        temp = get_products(i)
        all_products.append(temp)
    col = ['product_name', 'product_price', 'product_desc', 'product_img_url', 'product_buy_link']
    lazada_product_data = pd.DataFrame(all_products, columns=col)
    
    # Clean and convert product_price column to numeric for sorting
    lazada_product_data = lazada_product_data.sort_values(by='product_price', ascending=False)
    
    # Sort DataFrame by product_price in descending order
    lazada_product_data.to_csv(r'C:\\iGurus\\FlipKart (web Scrap)\\lazada_product_data(jeans).csv', encoding='utf-8', index=False)
```

```python
main('jeans')
```
